## 1. Write the Python code necessary to write your name with matplotlib lines.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

x=[-1 ,.5 ,.5,-1]
y=[ 0.5,  .25, -0.25, -1]

plt.plot(x,y, 'ro')

def connectpoints(x,y,p1,p2):
    x1, x2 = x[p1], x[p2]
    y1, y2 = y[p1], y[p2]
    plt.plot([x1,x2],[y1,y2],'k-')

connectpoints(x,y,0,1)
connectpoints(x,y,0,3)
connectpoints(x,y,1,2)
connectpoints(x,y,2,3)

x=[1.2 ,2 ,2,.5]
y=[ 0,  0, .5, -1]

connectpoints(x,y,0,1)
connectpoints(x,y,0,3)
connectpoints(x,y,1,2)
connectpoints(x,y,2,3)

plt.axis('equal')
plt.show()
```

## 2. Use matplotlib to plot the following equation: $y=x^2-x+2$. Add an anotation for the point 0, 0, the origin. 

```python
from matplotlib import pyplot as plt 
plt.figure(figsize=(10, 6))
x = list(range(-49, 50))
y = [n ** 2 - n + 2 for n in x]
plt.scatter(x, y)
plt.title('$y = x^2-x+2$')
plt.xlabel('$x$')
plt.ylabel('$x^2$')
plt.annotate('Origin', xy=(0,0), xytext=(-10,500), arrowprops={'facecolor': 'red'})
plt.show()
```

## 5. Comine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend.

```python
import math import sqrt, tan

plt.figure(figsize=(12, 10))

x = list(range(1, 5))

y1 = [math.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [math.tan(n) for n in x]
y4 = [2 ** n for n in x]

plt.plot(x,y1, c='red', label='$y=√x$')
plt.plot(x,y2, c='blue', label='$y=x^3$')
plt.plot(x,y3, c='green', label='$y=tan(x)$')
plt.plot(x,y4, c='orange', label='$y=2^x$')

plt.legend()

plt.show()
```
