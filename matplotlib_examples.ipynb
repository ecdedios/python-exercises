```python
from random import randint
```

```python
# some random data
x1 = [randint(-5, 25) + n for n in range(150)]
x2 = [x + randint(-6, 6) for x in x1]
```

```python
x1[0]
```

```python
plt.plot(x1, c='green', alpha=0.5)
plt.plot(x2, c='red', alpha=0.5)
plt.xlim(-20,200)
plt.ylim(-5, 160)
plt.show()
```

```python
plt.scatter(x = range(len(x1)), y = x1)
plt.scatter(x = range(len(x2)), y = x2)
plt.show()
```

```python
plt.scatter(range(len(x1)), x1, s=30, c='red', alpha=0.3)
plt.scatter(range(len(x1)), x1, s=10, c='green', alpha=0.4)
plt.show()
```

```python
plt.scatter(range(len(x1)), x1, s=30, c='pink', alpha=0.9)
plt.scatter(range(len(x1)), x1, s=10, c='purple', alpha=0.9)
plt.title('Random Generator')
plt.xlim(-20,200)
plt.ylim(-5, 160)
plt.show()
```

```python
x = list(range(-49,50))
y = [n ** 2 for n in x]
plt.scatter(x,y)
plt.title('My parabola')
plt.xlabel('$x^2$')
plt.ylabel('$x^2$')
plt.show()
```

```python
x = [1,2,3.14,4]
y = [n ** 2 for n in x]
plt.scatter(x,y)
plt.xticks([1,2,3.14,4], ['one', 'two', '$\pi$', 'four'])
plt.show()
```

```python
import math
```

```python
x = range(-4, 4)
y = [math.sin(n) for n in x]
plt.scatter(x, y)
plt.plot(x, y, ls='--')
plt.title('Approximation of the sine function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.yticks([1,2,3.14,4], ['one', 'two', 'three', 'four'], rotation=45)
plt.xticks(rotation=37)
plt.show()
```

```python
import random
weekdays = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'] * 2

x = range(14)
y = [random.randint(1,10) for _ in x]
plt.plot(x, y)
plt.xticks(x, weekdays, rotation=60)
plt.show()
```

```python
x = list(range(-4, 4))
x.append(3.14159)
y = [math.sin(n) for n in x]
plt.scatter(x, y)
plt.plot(x, y, ls=':')
plt.text(0.25,0, '(0,0)', fontsize=10, color='blue')
plt.text(1.2, 0.1, '($\pi$, sin($\pi$))', fontsize=14, color='red')

plt.show()
```

```python
x = range(-3, 10)
y = [n * 2 for n in x]
z = [n / 2 for n in x]

plt.plot(x, y, c='orange')
plt.plot(x, z, c='green')

plt.annotate('Intersection', xy=(0, 0), xytext=(-3, 5), arrowprops={'facecolor': 'blue'})
plt.show()
```

```python
x = range(-4, 5)
sin_x = [math.sin(n) for n in x]
cos_x = [math.cos(n) for n in x]

plt.figure(figsize=(14, 6)) # (width, height)

plt.scatter(x, sin_x, c='blue')
plt.scatter(x, cos_x, c='orange')
plt.plot(x, sin_x, c='blue', label='$\sin(x)$')
plt.plot(x, cos_x, c='orange', label='$\cos(x)$')

plt.legend(loc='upper right')
plt.title('Approximate sin and cosine curves')
plt.xlabel('$x$')

plt.show()
```
